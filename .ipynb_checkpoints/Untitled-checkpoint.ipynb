```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import os

# Set random seed for reproducibility
np.random.seed(42)

# Parameters
sequence_length = 50  # Number of time steps to look back
future_steps = 1     # Number of time steps to predict
train_ratio = 0.8    # Ratio of training data
batch_size = 32      # Batch size for training
epochs = 100         # Maximum number of epochs
patience = 10        # Patience for early stopping

def load_data(file_path):
    """Load data from CSV file"""
    try:
        df = pd.read_csv(file_path)
        print(f"Data loaded successfully. Shape: {df.shape}")
        print(f"Column names: {df.columns.tolist()}")
        print(f"First 5 rows:\n{df.head()}")
        print(f"Data statistics:\n{df.describe()}")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

def create_sequences(data, seq_length, future_steps):
    """Create sequences for LSTM input"""
    xs, ys = [], []
    for i in range(len(data) - seq_length - future_steps + 1):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length:(i + seq_length + future_steps)]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

def build_lstm_model(input_shape):
    """Build LSTM model architecture"""
    model = Sequential()
    model.add(LSTM(units=64, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(LSTM(units=64))
    model.add(Dropout(0.2))
    model.add(Dense(future_steps))
    model.compile(optimizer='adam', loss='mse', metrics=['mae'])
    print(model.summary())
    return model

def train_model(file_path='Trajectory2.csv'):
    """Main function to train LSTM model on trajectory data"""
    # Load data
    df = load_data(file_path)
    if df is None:
        return
    
    # Get the time series data
    data = df['x_Traject'].values.reshape(-1, 1)
    
    # Normalize data between 0 and 1
    scaler = MinMaxScaler(feature_range=(0, 1))
    data_scaled = scaler.fit_transform(data)
    
    # Create sequences
    X, y = create_sequences(data_scaled, sequence_length, future_steps)
    
    # Reshape X to [samples, time steps, features]
    X = X.reshape(X.shape[0], X.shape[1], 1)
    
    # Split into train and validation sets
    train_size = int(len(X) * train_ratio)
    X_train, X_val = X[:train_size], X[train_size:]
    y_train, y_val = y[:train_size], y[train_size:]
    
    print(f"Training data shape: {X_train.shape}, {y_train.shape}")
    print(f"Validation data shape: {X_val.shape}, {y_val.shape}")
    
    # Build model
    model = build_lstm_model((X_train.shape[1], 1))
    
    # Early stopping callback
    early_stopping = EarlyStopping(
        monitor='val_loss',
        patience=patience,
        restore_best_weights=True
    )
    
    # Train the model
    history = model.fit(
        X_train, y_train,
        epochs=epochs,
        batch_size=batch_size,
        validation_data=(X_val, y_val),
        callbacks=[early_stopping],
        verbose=1
    )
    
    # Evaluate the model
    train_loss = model.evaluate(X_train, y_train, verbose=0)
    val_loss = model.evaluate(X_val, y_val, verbose=0)
    print(f"Training Loss: {train_loss[0]:.4f}, Training MAE: {train_loss[1]:.4f}")
    print(f"Validation Loss: {val_loss[0]:.4f}, Validation MAE: {val_loss[1]:.4f}")
    
    # Make predictions
    train_predict = model.predict(X_train)
    val_predict = model.predict(X_val)
    
    # Ensure arrays are 2D for inverse_transform
    train_predict_2d = train_predict.reshape(train_predict.shape[0], -1)
    val_predict_2d = val_predict.reshape(val_predict.shape[0], -1)
    y_train_2d = y_train.reshape(y_train.shape[0], -1)
    y_val_2d = y_val.reshape(y_val.shape[0], -1)
    
    # Inverse transform predictions to original scale
    train_predict = scaler.inverse_transform(train_predict_2d)
    val_predict = scaler.inverse_transform(val_predict_2d)
    y_train_inv = scaler.inverse_transform(y_train_2d)
    y_val_inv = scaler.inverse_transform(y_val_2d)
    
    # Calculate RMSE
    train_rmse = np.sqrt(mean_squared_error(y_train_inv, train_predict))
    val_rmse = np.sqrt(mean_squared_error(y_val_inv, val_predict))
    print(f"Training RMSE: {train_rmse:.4f}")
    print(f"Validation RMSE: {val_rmse:.4f}")
    
    # Plot training & validation loss
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Model Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot predictions vs actual
    plt.subplot(1, 2, 2)
    plt.figure(figsize=(12, 6))
    
    # Plot a subset of predictions for better visibility
    subset_size = min(500, len(y_val_inv))
    
    # Create timestamps for x-axis
    timestamps = np.arange(subset_size)
    
    plt.plot(timestamps, y_val_inv[:subset_size], label='Actual')
    plt.plot(timestamps, val_predict[:subset_size], label='Predicted')
    plt.title('Prediction vs Actual')
    plt.xlabel('Time Steps')
    plt.ylabel('Trajectory Value')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('lstm_prediction_results.png')
    plt.show()
    
    # Save model
    model.save('lstm_trajectory_model.h5')
    print("Model saved as 'lstm_trajectory_model.h5'")
    
    return model, scaler

def predict_future(model, scaler, data, steps_ahead=100):
    """Predict future values using the trained model"""
    # Use the last sequence_length values for prediction
    last_sequence = data[-sequence_length:].reshape(1, sequence_length, 1)
    
    future_predictions = []
    current_sequence = last_sequence.copy()
    
    # Predict one step at a time and update the sequence
    for _ in range(steps_ahead):
        # Predict the next value
        next_value = model.predict(current_sequence, verbose=0)
        future_predictions.append(next_value[0, 0])
        
        # Update the sequence by removing the first value and appending the prediction
        current_sequence = np.roll(current_sequence, -1, axis=1)
        current_sequence[0, -1, 0] = next_value[0, 0]
    
    # Inverse transform to original scale
    future_predictions = np.array(future_predictions).reshape(-1, 1)
    future_predictions = scaler.inverse_transform(future_predictions)
    
    return future_predictions

if __name__ == "__main__":
    # Train the model
    model, scaler = train_model()
    
    # Load the data again for future prediction
    df = load_data('Trajectory2.csv')
    data = df['x_Traject'].values.reshape(-1, 1)
    data_scaled = scaler.transform(data)
    
    # Predict future values
    future_values = predict_future(model, scaler, data_scaled, steps_ahead=100)
    
    # Plot original data and future predictions
    plt.figure(figsize=(12, 6))
    
    # Plot original data
    plt.plot(np.arange(len(data)), data, label='Historical Data')
    
    # Plot future predictions
    plt.plot(np.arange(len(data), len(data) + len(future_values)), future_values, 
             color='red', label='Future Predictions')
    
    plt.title('Trajectory Prediction')
    plt.xlabel('Time Steps')
    plt.ylabel('Trajectory Value')
    plt.legend()
    plt.grid(True)
    plt.savefig('future_predictions.png')
    plt.show()
    
    print("Prediction completed and visualized in 'future_predictions.png'")
```

```python
# Copy and paste this entire cell into your Jupyter notebook
# Make sure to run it after your model is trained

%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Set larger figure size for better notebook display
plt.rcParams["figure.figsize"] = (18, 12)
sns.set_style("whitegrid")

# Helper function for sequences
def create_sequences(data, seq_length, future_steps):
    xs, ys = [], []
    for i in range(len(data) - seq_length - future_steps + 1):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length:(i + seq_length + future_steps)]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

# Load data
df = pd.read_csv('Trajectory2.csv')
data = df['x_Traject'].values.reshape(-1, 1)

# Get predictions from model (using the sequence_length from your training)
sequence_length = 50  # Update this if you used a different value
data_scaled = scaler.transform(data)
X, y = create_sequences(data_scaled, sequence_length, 1)
X = X.reshape(X.shape[0], X.shape[1], 1)
predictions = model.predict(X, verbose=0)

# Reshape for inverse transform
y_inv = scaler.inverse_transform(y.reshape(y.shape[0], -1))
pred_inv = scaler.inverse_transform(predictions.reshape(predictions.shape[0], -1))

# Calculate metrics
rmse = np.sqrt(mean_squared_error(y_inv, pred_inv))
mae = mean_absolute_error(y_inv, pred_inv)
r2 = r2_score(y_inv, pred_inv)
corr = np.corrcoef(y_inv.flatten(), pred_inv.flatten())[0, 1]

# Calculate and normalize errors
errors = y_inv - pred_inv
abs_errors = np.abs(errors)
mean_error = np.mean(abs_errors)
max_error = np.max(abs_errors)

# Detect anomalies (points with unusually high errors)
error_threshold = np.mean(abs_errors) + 2.5 * np.std(abs_errors)
anomaly_indices = np.where(abs_errors > error_threshold)[0]

# Create a figure with 2x3 subplots
fig, axes = plt.subplots(3, 2, figsize=(18, 18))
fig.suptitle('LSTM Model Analysis Dashboard', fontsize=24, fontweight='bold', y=0.98)

# 1. Original vs Predicted (last 500 points)
ax1 = axes[0, 0]
window_size = 500
start_idx = max(0, len(data) - window_size)

# Plot original data
ax1.plot(range(window_size), data[start_idx:start_idx+window_size], 
         'b-', linewidth=1.5, alpha=0.7, label='Original')

# Plot predictions
pred_offset = sequence_length  # Adjust based on your model
pred_range = range(pred_offset, window_size)
ax1.plot(pred_range, 
         pred_inv[start_idx-sequence_length:start_idx-sequence_length+window_size-pred_offset].flatten(), 
         'r-', linewidth=1.5, alpha=0.7, label='Predicted')

# Highlight anomalies
anomalies_in_range = [idx for idx in anomaly_indices 
                     if start_idx-sequence_length <= idx < start_idx-sequence_length+window_size-pred_offset]
if anomalies_in_range:
    anomaly_x = [idx - (start_idx-sequence_length) + pred_offset for idx in anomalies_in_range]
    anomaly_y = [pred_inv[idx][0] for idx in anomalies_in_range]
    ax1.scatter(anomaly_x, anomaly_y, color='yellow', edgecolor='black', s=80, 
                zorder=5, alpha=0.8, label='Anomalies')

ax1.set_title('Original vs Predicted Trajectory (Last 500 Points)', fontsize=16)
ax1.set_xlabel('Time Steps', fontsize=12)
ax1.set_ylabel('Value', fontsize=12)
ax1.legend(fontsize=12)
ax1.grid(True, alpha=0.3)

# 2. Error Analysis
ax2 = axes[0, 1]
sns.histplot(errors.flatten(), kde=True, ax=ax2, bins=30, color='purple', alpha=0.6)
ax2.axvline(x=0, color='red', linestyle='--', alpha=0.6)
ax2.set_title('Prediction Error Distribution', fontsize=16)
ax2.set_xlabel('Error Value', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)

# Add statistics text
stats_text = (f"Mean Error: {np.mean(errors):.4f}\n"
              f"Mean Abs Error: {mean_error:.4f}\n"
              f"Error Std Dev: {np.std(errors):.4f}\n"
              f"Max Error: {max_error:.4f}")
ax2.text(0.95, 0.95, stats_text, transform=ax2.transAxes, fontsize=12,
         verticalalignment='top', horizontalalignment='right',
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 3. Future Predictions with Confidence Intervals
ax3 = axes[1, 0]
steps_ahead = 100
history_window = 100

# Generate multiple future prediction scenarios for confidence intervals
n_scenarios = 20
future_predictions = np.zeros((n_scenarios, steps_ahead))

for s in range(n_scenarios):
    # Get last sequence from the data
    last_sequence = data_scaled[-sequence_length:].reshape(1, sequence_length, 1)
    
    # Add small noise for Monte Carlo simulation (except first run)
    if s > 0:
        noise_level = 0.01
        last_sequence = last_sequence + np.random.normal(0, noise_level, last_sequence.shape)
    
    # Generate future steps
    current_sequence = last_sequence.copy()
    for i in range(steps_ahead):
        # Predict next step
        next_pred = model.predict(current_sequence, verbose=0)[0][0]
        future_predictions[s, i] = next_pred
        
        # Update sequence
        current_sequence = np.roll(current_sequence, -1, axis=1)
        current_sequence[0, -1, 0] = next_pred

# Calculate mean and confidence intervals
future_mean = np.mean(future_predictions, axis=0)
future_upper = np.percentile(future_predictions, 95, axis=0)
future_lower = np.percentile(future_predictions, 5, axis=0)

# Transform to original scale
future_mean = scaler.inverse_transform(future_mean.reshape(-1, 1)).flatten()
future_upper = scaler.inverse_transform(future_upper.reshape(-1, 1)).flatten()
future_lower = scaler.inverse_transform(future_lower.reshape(-1, 1)).flatten()

# Plot historical data
historical = data[-history_window:].flatten()
ax3.plot(range(history_window), historical, 'b-', linewidth=1.5, label='Historical')

# Plot future predictions with confidence interval
ax3.plot(range(history_window-1, history_window+steps_ahead), 
         np.concatenate([historical[-1:], future_mean]), 
         'r-', linewidth=1.5, label='Forecast')

# Add confidence intervals
ax3.fill_between(range(history_window-1, history_window+steps_ahead),
                 np.concatenate([historical[-1:], future_lower]),
                 np.concatenate([historical[-1:], future_upper]),
                 color='red', alpha=0.2, label='90% Confidence')

# Mark present
ax3.axvline(x=history_window-1, color='black', linestyle='--', label='Present')

ax3.set_title('Future Trajectory Forecast with Confidence Interval', fontsize=16)
ax3.set_xlabel('Time Steps', fontsize=12)
ax3.set_ylabel('Value', fontsize=12)
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3)

# 4. Actual vs Predicted Scatter
ax4 = axes[1, 1]
scatter = ax4.scatter(y_inv.flatten(), pred_inv.flatten(), 
                     c=abs_errors.flatten(), cmap='viridis', 
                     alpha=0.6, s=20)

# Add perfect prediction line
min_val = min(np.min(y_inv), np.min(pred_inv))
max_val = max(np.max(y_inv), np.max(pred_inv))
ax4.plot([min_val, max_val], [min_val, max_val], 'r--')

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax4)
cbar.set_label('Absolute Error')

ax4.set_title('Actual vs Predicted Values', fontsize=16)
ax4.set_xlabel('Actual Values', fontsize=12)
ax4.set_ylabel('Predicted Values', fontsize=12)

# Add R² annotation
ax4.text(0.05, 0.95, f"R² = {r2:.4f}", transform=ax4.transAxes,
        fontsize=12, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# 5. Error Over Time
ax5 = axes[2, 0]
x_indices = np.arange(len(errors))

# Using a stem plot for error visualization
markerline, stemlines, baseline = ax5.stem(
    x_indices[::50],  # Sample every 50th point for clearer visualization
    abs_errors.flatten()[::50],
    linefmt='grey', markerfmt='o', basefmt='grey')

plt.setp(stemlines, 'linewidth', 0.8, 'alpha', 0.6)
plt.setp(markerline, 'markersize', 4, 'color', 'blue', 'alpha', 0.8)

# Add threshold line
ax5.axhline(y=error_threshold, color='red', linestyle='--', 
           alpha=0.7, label=f'Anomaly Threshold ({error_threshold:.4f})')

ax5.set_title('Error Magnitude Over Time (Sampled)', fontsize=16)
ax5.set_xlabel('Data Point Index', fontsize=12)
ax5.set_ylabel('Absolute Error', fontsize=12)
ax5.legend(fontsize=10)
ax5.grid(True, alpha=0.3)

# 6. Key Performance Indicators
ax6 = axes[2, 1]
ax6.axis('off')

# Create KPI section with color-coding
kpi_text = "KEY PERFORMANCE INDICATORS\n\n"

# RMSE
data_range = np.max(data) - np.min(data)
rmse_relative = rmse / data_range
if rmse_relative < 0.05:
    rmse_color = 'green'
elif rmse_relative < 0.1:
    rmse_color = 'orange'
else:
    rmse_color = 'red'
kpi_text += f"RMSE: {rmse:.4f} "
kpi_text += f"({rmse_relative*100:.1f}% of range) - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if rmse_relative > 0.1 else 1 if rmse_relative > 0.05 else 2]}\n\n"

# MAE
mae_relative = mae / data_range
if mae_relative < 0.03:
    mae_color = 'green'
elif mae_relative < 0.08:
    mae_color = 'orange'
else:
    mae_color = 'red'
kpi_text += f"MAE: {mae:.4f} "
kpi_text += f"({mae_relative*100:.1f}% of range) - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if mae_relative > 0.08 else 1 if mae_relative > 0.03 else 2]}\n\n"

# R² Score
if r2 > 0.9:
    r2_color = 'green'
elif r2 > 0.8:
    r2_color = 'orange'
else:
    r2_color = 'red'
kpi_text += f"R² Score: {r2:.4f} - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if r2 < 0.8 else 1 if r2 < 0.9 else 2]}\n\n"

# Correlation
if corr > 0.95:
    corr_color = 'green'
elif corr > 0.85:
    corr_color = 'orange'
else:
    corr_color = 'red'
kpi_text += f"Correlation: {corr:.4f} - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if corr < 0.85 else 1 if corr < 0.95 else 2]}\n\n"

# Anomaly percentage
anomaly_pct = len(anomaly_indices) / len(y_inv) * 100
if anomaly_pct < 2:
    anomaly_color = 'green'
elif anomaly_pct < 5:
    anomaly_color = 'orange'
else:
    anomaly_color = 'red'
kpi_text += f"Anomalies: {anomaly_pct:.2f}% of points - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if anomaly_pct > 5 else 1 if anomaly_pct > 2 else 2]}\n\n"

# Future prediction confidence (CI width)
ci_width = np.mean(future_upper - future_lower) / np.mean(future_mean) * 100
if ci_width < 10:
    ci_color = 'green'
elif ci_width < 20:
    ci_color = 'orange'
else:
    ci_color = 'red'
kpi_text += f"Forecast Confidence: ±{ci_width:.1f}% - "
kpi_text += f"Quality: {['Poor', 'Average', 'Good'][0 if ci_width > 20 else 1 if ci_width > 10 else 2]}\n\n"

# Add model insights
kpi_text += "\nMODEL INSIGHTS & RECOMMENDATIONS:\n\n"

if r2 < 0.9:
    kpi_text += "• Consider increasing model complexity\n"
if anomaly_pct > 2:
    kpi_text += "• Investigate anomalous predictions\n"
if ci_width > 15:
    kpi_text += "• Future predictions have high uncertainty\n"
if rmse_relative > 0.05:
    kpi_text += "• Error rate suggests room for improvement\n"
if corr < 0.95:
    kpi_text += "• Model may miss some pattern details\n"

ax6.text(0.5, 0.5, kpi_text, 
        horizontalalignment='center', verticalalignment='center',
        fontsize=12, wrap=True, transform=ax6.transAxes,
        bbox=dict(boxstyle='round,pad=1', facecolor='#f0f0f0', edgecolor='#cccccc', alpha=0.9))

plt.tight_layout()
plt.subplots_adjust(top=0.93)

# Display the plot in the notebook
plt.savefig('lstm_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"Dashboard saved as 'lstm_dashboard.png'")
print(f"RMSE: {rmse:.4f}, MAE: {mae:.4f}, R²: {r2:.4f}")
```

```python
# Load and visualize the saved LSTM model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import load_model
import os

# Parameters (same as used in training)
sequence_length = 50  # Number of time steps to look back
future_steps = 1     # Number of time steps to predict
train_ratio = 0.8    # Ratio of training data

# Functions from original code
def load_data(file_path):
    """Load data from CSV file"""
    try:
        df = pd.read_csv(file_path)
        print(f"Data loaded successfully. Shape: {df.shape}")
        print(f"Column names: {df.columns.tolist()}")
        print(f"First 5 rows:\n{df.head()}")
        print(f"Data statistics:\n{df.describe()}")
        return df
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

def create_sequences(data, seq_length, future_steps):
    """Create sequences for LSTM input"""
    xs, ys = [], []
    for i in range(len(data) - seq_length - future_steps + 1):
        x = data[i:(i + seq_length)]
        y = data[i + seq_length:(i + seq_length + future_steps)]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

def predict_future(model, scaler, data, steps_ahead=100):
    """Predict future values using the trained model"""
    # Use the last sequence_length values for prediction
    last_sequence = data[-sequence_length:].reshape(1, sequence_length, 1)
    
    future_predictions = []
    current_sequence = last_sequence.copy()
    
    # Predict one step at a time and update the sequence
    for _ in range(steps_ahead):
        # Predict the next value
        next_value = model.predict(current_sequence, verbose=0)
        future_predictions.append(next_value[0, 0])
        
        # Update the sequence by removing the first value and appending the prediction
        current_sequence = np.roll(current_sequence, -1, axis=1)
        current_sequence[0, -1, 0] = next_value[0, 0]
    
    # Inverse transform to original scale
    future_predictions = np.array(future_predictions).reshape(-1, 1)
    future_predictions = scaler.inverse_transform(future_predictions)
    
    return future_predictions

# Load the model
print("Loading saved model...")
model_path = 'lstm_trajectory_model.h5'
try:
    model = load_model(model_path)
    print(f"Model loaded successfully from {model_path}")
    # Print model summary
    model.summary()
except Exception as e:
    print(f"Error loading model: {e}")
    print("Please ensure the model file exists and is in the correct format.")
    # Uncomment to exit if model loading fails
    # import sys; sys.exit(1)

# Load the data
print("\nLoading data...")
df = load_data('Trajectory2.csv')
if df is None:
    print("Error: Could not load data.")
    # Uncomment to exit if data loading fails
    # import sys; sys.exit(1)

# Get the time series data and normalize
data = df['x_Traject'].values.reshape(-1, 1)
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Create sequences
X, y = create_sequences(data_scaled, sequence_length, future_steps)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split into train and validation sets (to match original training split)
train_size = int(len(X) * train_ratio)
X_train, X_val = X[:train_size], X[train_size:]
y_train, y_val = y[:train_size], y[train_size:]

print(f"Processing data - shapes: X_train: {X_train.shape}, X_val: {X_val.shape}")

# Generate predictions
print("Generating predictions...")
train_predict = model.predict(X_train, verbose=0)
val_predict = model.predict(X_val, verbose=0)

# Reshape for inverse transform
train_predict_2d = train_predict.reshape(train_predict.shape[0], -1)
val_predict_2d = val_predict.reshape(val_predict.shape[0], -1)
y_train_2d = y_train.reshape(y_train.shape[0], -1)
y_val_2d = y_val.reshape(y_val.shape[0], -1)

# Inverse transform to original scale
train_predict = scaler.inverse_transform(train_predict_2d)
val_predict = scaler.inverse_transform(val_predict_2d)
y_train_inv = scaler.inverse_transform(y_train_2d)
y_val_inv = scaler.inverse_transform(y_val_2d)

# Calculate metrics
train_rmse = np.sqrt(mean_squared_error(y_train_inv, train_predict))
val_rmse = np.sqrt(mean_squared_error(y_val_inv, val_predict))
train_mae = mean_absolute_error(y_train_inv, train_predict)
val_mae = mean_absolute_error(y_val_inv, val_predict)

print(f"Model Performance Metrics:")
print(f"Training RMSE: {train_rmse:.4f}")
print(f"Validation RMSE: {val_rmse:.4f}")
print(f"Training MAE: {train_mae:.4f}")
print(f"Validation MAE: {val_mae:.4f}")

# Calculate residuals
train_residuals = y_train_inv - train_predict
val_residuals = y_val_inv - val_predict

# Predict future values
print("Predicting future values...")
future_values = predict_future(model, scaler, data_scaled, steps_ahead=100)
print(f"Generated {len(future_values)} future predictions")

# Create comprehensive visualization dashboard
print("\nCreating visualizations...")

# Set basic plot parameters
plt.rcParams.update({
    'figure.figsize': (20, 14),
    'axes.grid': True,
    'grid.alpha': 0.7,
    'grid.linestyle': '--',
    'axes.titlesize': 14,
    'axes.titleweight': 'bold',
    'axes.labelsize': 12,
    'legend.fontsize': 10
})

# Main dashboard visualization
fig = plt.figure(figsize=(20, 14))
gs = GridSpec(3, 2, figure=fig)

# 1. Validation Predictions vs Actual (Full set)
ax1 = fig.add_subplot(gs[0, 0])
subset_size = min(500, len(y_val_inv))
timestamps = np.arange(subset_size)
ax1.plot(timestamps, y_val_inv[:subset_size], 'b-', linewidth=1.5, label='Actual Values')
ax1.plot(timestamps, val_predict[:subset_size], 'r-', linewidth=1.5, label='Predicted Values')
ax1.set_title('Validation: Predicted vs Actual (First 500 points)')
ax1.set_xlabel('Time Steps')
ax1.set_ylabel('Trajectory Value')
ax1.legend(loc='best')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.annotate(f'RMSE: {val_rmse:.4f}\nMAE: {val_mae:.4f}', 
             xy=(0.05, 0.90), xycoords='axes fraction', fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

# 2. Zoomed-in Validation Prediction
ax2 = fig.add_subplot(gs[0, 1])
zoom_start = min(100, len(y_val_inv) // 2)
zoom_size = min(100, len(y_val_inv) - zoom_start)
zoom_timestamps = np.arange(zoom_start, zoom_start + zoom_size)
ax2.plot(zoom_timestamps, y_val_inv[zoom_start:zoom_start + zoom_size], 'b-', linewidth=2, label='Actual')
ax2.plot(zoom_timestamps, val_predict[zoom_start:zoom_start + zoom_size], 'r-', linewidth=2, label='Predicted')
ax2.set_title(f'Zoomed Validation Predictions (Steps {zoom_start}-{zoom_start+zoom_size})')
ax2.set_xlabel('Time Steps')
ax2.set_ylabel('Trajectory Value')
ax2.legend(loc='best')
ax2.grid(True, linestyle='--', alpha=0.7)

# 3. Residuals Plot (Validation)
ax3 = fig.add_subplot(gs[1, 0])
ax3.scatter(val_predict.flatten(), val_residuals.flatten(), alpha=0.5, c='blue')
ax3.axhline(y=0, color='r', linestyle='-', linewidth=1)
ax3.set_title('Validation Residuals')
ax3.set_xlabel('Predicted Values')
ax3.set_ylabel('Residuals (Actual - Predicted)')
ax3.grid(True, linestyle='--', alpha=0.7)

# 4. Distribution of Residuals
ax4 = fig.add_subplot(gs[1, 1])
sns.histplot(val_residuals.flatten(), kde=True, ax=ax4, color='blue', alpha=0.7)
ax4.axvline(x=0, color='r', linestyle='-', linewidth=1)
ax4.set_title('Distribution of Validation Residuals')
ax4.set_xlabel('Residual Value')
ax4.set_ylabel('Frequency')
ax4.grid(True, linestyle='--', alpha=0.7)
# Add statistics on the plot
ax4.annotate(f'Mean: {np.mean(val_residuals):.4f}\nStd: {np.std(val_residuals):.4f}',
             xy=(0.05, 0.90), xycoords='axes fraction', fontsize=10,
             bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

# 5. Original Data and Future Predictions
ax5 = fig.add_subplot(gs[2, :])
# Plot original data
ax5.plot(np.arange(len(data)), data, 'b-', linewidth=1.5, label='Historical Data')
# Plot future predictions
ax5.plot(np.arange(len(data), len(data) + len(future_values)), future_values, 
         'r-', linewidth=1.5, label='Future Predictions (100 steps)')
# Add a vertical line to separate historical data from predictions
ax5.axvline(x=len(data), color='k', linestyle='--', linewidth=1)
ax5.set_title('Trajectory Prediction with Future Forecast')
ax5.set_xlabel('Time Steps')
ax5.set_ylabel('Trajectory Value')
ax5.legend(loc='best')
ax5.grid(True, linestyle='--', alpha=0.7)

# Add a text box with model information and metrics
model_info = (
    f"Model: LSTM\n"
    f"Sequence Length: {sequence_length}\n"
    f"Training RMSE: {train_rmse:.4f}\n"
    f"Validation RMSE: {val_rmse:.4f}\n"
    f"Training MAE: {train_mae:.4f}\n"
    f"Validation MAE: {val_mae:.4f}"
)
fig.text(0.5, 0.01, model_info, ha='center', va='bottom', 
         bbox=dict(boxstyle="round,pad=0.5", fc="white", ec="gray", alpha=0.8),
         fontsize=12)

plt.suptitle('LSTM Model Performance Analysis', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.98])
plt.show()

# Additional detailed visualizations

# 1. Detailed view of predictions in different segments
plt.figure(figsize=(18, 12))

# Plot four different segments of the validation data
segment_length = min(100, len(y_val_inv)//4)
for i in range(4):
    start_idx = i * segment_length
    end_idx = min(start_idx + segment_length, len(y_val_inv))
    
    plt.subplot(2, 2, i+1)
    plt.plot(np.arange(start_idx, end_idx), y_val_inv[start_idx:end_idx], 'b-', label='Actual')
    plt.plot(np.arange(start_idx, end_idx), val_predict[start_idx:end_idx], 'r-', label='Predicted')
    plt.title(f'Segment {i+1}: Steps {start_idx}-{end_idx-1}')
    plt.xlabel('Time Steps')
    plt.ylabel('Trajectory Value')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Add RMSE for this segment
    segment_rmse = np.sqrt(mean_squared_error(y_val_inv[start_idx:end_idx], val_predict[start_idx:end_idx]))
    plt.annotate(f'RMSE: {segment_rmse:.4f}', 
                xy=(0.05, 0.90), xycoords='axes fraction', fontsize=10,
                bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))

plt.suptitle('Validation Predictions - Multiple Segments', fontsize=16, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# 2. Error analysis across the sequence
plt.figure(figsize=(12, 8))

# Calculate absolute errors for validation set
abs_errors = np.abs(y_val_inv - val_predict)

# Plot the absolute errors
plt.subplot(2, 1, 1)
plt.plot(np.arange(len(abs_errors)), abs_errors, 'r-', linewidth=1)
plt.axhline(y=np.mean(abs_errors), color='k', linestyle='--', label=f'Mean: {np.mean(abs_errors):.4f}')
plt.title('Absolute Error Over Validation Sequence')
plt.xlabel('Sequence Index')
plt.ylabel('Absolute Error')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Rolling average of errors
window = 20
if len(abs_errors) > window:
    rolling_mean = np.convolve(abs_errors.flatten(), np.ones(window)/window, mode='valid')
    plt.subplot(2, 1, 2)
    plt.plot(np.arange(len(rolling_mean)), rolling_mean, 'g-', linewidth=1.5)
    plt.title(f'Rolling Average of Absolute Error (Window Size: {window})')
    plt.xlabel('Sequence Index')
    plt.ylabel('Rolling Average Error')
    plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# 3. Future prediction with confidence intervals (estimated based on model error)
plt.figure(figsize=(14, 7))

# Calculate the standard deviation of validation residuals
std_residuals = np.std(val_residuals)

# Plot the original data
plt.plot(np.arange(len(data)), data, 'b-', linewidth=1.5, label='Historical Data')

# Plot future predictions with confidence intervals
future_range = np.arange(len(data), len(data) + len(future_values))
plt.plot(future_range, future_values, 'r-', linewidth=2, label='Future Predictions')

# Add confidence intervals (±1 and ±2 standard deviations)
plt.fill_between(future_range, 
                 (future_values - std_residuals).flatten(), 
                 (future_values + std_residuals).flatten(), 
                 color='red', alpha=0.3, label='68% Confidence Interval')
plt.fill_between(future_range, 
                 (future_values - 2*std_residuals).flatten(), 
                 (future_values + 2*std_residuals).flatten(), 
                 color='red', alpha=0.1, label='95% Confidence Interval')

# Add a vertical line to separate historical data from predictions
plt.axvline(x=len(data), color='k', linestyle='--', linewidth=1)
plt.title('Trajectory Prediction with Confidence Intervals', fontsize=14, fontweight='bold')
plt.xlabel('Time Steps', fontsize=12)
plt.ylabel('Trajectory Value', fontsize=12)
plt.legend(loc='best')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("Visualization completed successfully!")

# Optional: Interactive exploration of validation data
try:
    from ipywidgets import interact, IntSlider

    def plot_window(start_idx, window_size=100):
        """Plot a specific window of the validation predictions"""
        plt.figure(figsize=(12, 6))
        end_idx = min(start_idx + window_size, len(val_predict))
        x_vals = np.arange(start_idx, end_idx)
        
        plt.plot(x_vals, y_val_inv[start_idx:end_idx], 'b-', linewidth=2, label='Actual')
        plt.plot(x_vals, val_predict[start_idx:end_idx], 'r-', linewidth=2, label='Predicted')
        
        plt.title(f'Validation Predictions (Steps {start_idx} to {end_idx-1})', fontsize=14)
        plt.xlabel('Time Steps', fontsize=12)
        plt.ylabel('Trajectory Value', fontsize=12)
        plt.legend(loc='best')
        plt.grid(True, linestyle='--', alpha=0.7)
        
        # Add RMSE for this window
        window_rmse = np.sqrt(mean_squared_error(y_val_inv[start_idx:end_idx], val_predict[start_idx:end_idx]))
        plt.annotate(f'Window RMSE: {window_rmse:.4f}', 
                    xy=(0.05, 0.90), xycoords='axes fraction', fontsize=10,
                    bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.8))
        
        plt.tight_layout()
        plt.show()

    # Create an interactive slider to explore different time windows
    max_idx = len(val_predict) - 100  # Ensuring we have at least 100 points to show
    if max_idx > 0:
        print("\nUse the sliders below to explore different time windows:")
        interact(
            plot_window,
            start_idx=IntSlider(min=0, max=max_idx, step=10, value=0, description='Start Index:'),
            window_size=IntSlider(min=50, max=500, step=50, value=100, description='Window Size:')
        )
    else:
        print("Not enough validation data for interactive visualization")
except ImportError:
    print("Interactive widgets not available. Install ipywidgets for interactive exploration.")
```
